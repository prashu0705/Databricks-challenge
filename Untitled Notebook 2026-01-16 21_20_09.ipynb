```python
from pyspark.sql.types import *
from pyspark.sql import functions as F

schema = StructType([
    StructField("event_time", TimestampType(), True),
    StructField("event_type", StringType(), True),
    StructField("product_id", LongType(), True),
    StructField("category_id", LongType(), True),
    StructField("category_code", StringType(), True),
    StructField("brand", StringType(), True),
    StructField("price", DoubleType(), True),
    StructField("user_id", LongType(), True),
    StructField("user_session", StringType(), True)
])
```

```python
df_oct = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Oct.csv",
    header=True,
    schema=schema
)
```

```python
df_nov = spark.read.csv(
    "/Volumes/workspace/ecommerce/ecommerce_data/2019-Nov.csv",
    header=True,
    schema=schema
)
```

```python
events = df_oct.unionByName(df_nov)
events.printSchema()
events.show(5, truncate=False)
```

```python
bronze = events.withColumn("ingestion_ts", F.current_timestamp())

bronze.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.ecommerce.bronze_events")
```

```python
silver = spark.table("workspace.ecommerce.bronze_events") \
    .filter(F.col("price") >= 0) \
    .dropDuplicates(["event_time", "user_session", "product_id"])

silver.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.ecommerce.silver_events")
```

```python
gold = spark.table("workspace.ecommerce.silver_events") \
    .filter(F.col("event_type") == "purchase") \
    .groupBy("product_id") \
    .agg(
        F.count("*").alias("total_purchases"),
        F.sum("price").alias("total_revenue")
    )

gold.write.format("delta") \
    .mode("overwrite") \
    .saveAsTable("workspace.ecommerce.gold_product_metrics")
```

```python
dbutils.widgets.dropdown(
    "layer",
    "bronze",
    ["bronze", "silver", "gold"]
)

layer = dbutils.widgets.get("layer")
print("Running layer:", layer)
```

```python
if layer == "bronze":
    events.withColumn("ingestion_ts", F.current_timestamp()) \
        .write.format("delta") \
        .mode("overwrite") \
        .saveAsTable("workspace.ecommerce.bronze_events")

elif layer == "silver":
    spark.table("workspace.ecommerce.bronze_events") \
        .filter(F.col("price") >= 0) \
        .dropDuplicates(["event_time", "user_session", "product_id"]) \
        .write.format("delta") \
        .mode("overwrite") \
        .saveAsTable("workspace.ecommerce.silver_events")

elif layer == "gold":
    spark.table("workspace.ecommerce.silver_events") \
        .filter(F.col("event_type") == "purchase") \
        .groupBy("product_id") \
        .agg(
            F.count("*").alias("total_purchases"),
            F.sum("price").alias("total_revenue")
        ) \
        .write.format("delta") \
        .mode("overwrite") \
        .saveAsTable("workspace.ecommerce.gold_product_metrics")
```

```python
%sql
CREATE VIEW workspace.ecommerce.top_products AS
SELECT
  product_id,
  total_purchases,
  total_revenue
FROM workspace.ecommerce.gold_product_metrics
WHERE total_purchases > 10
ORDER BY total_revenue DESC;
```

```python
%sql
SELECT * FROM workspace.ecommerce.top_products LIMIT 20;
```

```python
%sql
CREATE CATALOG ecommerce;
USE CATALOG ecommerce;
CREATE SCHEMA bronze;
CREATE SCHEMA silver;
CREATE SCHEMA gold;
```

```python
%sql
CREATE TABLE bronze.events USING DELTA;
CREATE TABLE silver.events USING DELTA;
CREATE TABLE gold.products USING DELTA;
```

```python
%sql
CREATE OR REPLACE TABLE gold.products AS
SELECT product_id,
       COUNT(*) AS total_purchases,
       SUM(price) AS total_revenue
FROM workspace.ecommerce.silver_events
WHERE event_type = 'purchase'
GROUP BY product_id;
```

```python
%sql
SELECT product_id, total_purchases, total_revenue
FROM gold.products
ORDER BY total_revenue DESC
LIMIT 100;
```
